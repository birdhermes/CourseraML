```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
import math
```

```python
#генерирование выборки объемом 1000 по распеределению хи-квадрат

x = np.linspace(0,25,1000)
k=5
rv = sts.chi2(k)
pdf = rv.pdf(x)


y=rv.rvs(1000)
pdf = rv.pdf(x)
plt.plot(x, pdf)
plt.hist(y,normed=True)
plt.ylabel('y')
plt.xlabel('x')
plt.title('гистограмма выборки объемом 1000 по распеределению хи-квадрат')
```

```python
# в функции выполняется генерации списка для выборки со значением объема n для выборочных средних 
# в функции передается объем выборки n
def func_list(n):
    X_array =[]
    for i in range(1000):
        result=rv.rvs(n)
        X_array.append(result.mean())
    return(X_array)
```

```python
#гистограмма выборки 5  и плосность нормального распределения с нужными параметрами  
sample5=func_list(5)
sigma=np.sqrt(2*k/5)
plt.plot(x,pdf)
plt.hist(sample5,normed=True)
plt.ylabel('y')
plt.xlabel('x')
plt.title('распределение выборочного среднего случайной величины при объём выборки n=5')
```

```python
#гистограмма выборки 10 и плосность нормального распределения с нужными параметрами  
sample10=func_list(10)
sigma=np.sqrt(2*k/10)
gucci=sts.norm(loc=k,scale=sigma)
pdf=gucci.pdf(x)
plt.plot(x,pdf)
plt.hist(sample10,normed=True)
plt.ylabel('y')
plt.xlabel('x')
plt.title('распределение выборочного среднего случайной величины при объёме выборки n=10')
```

```python
#гистограмма выборки 50 и плосность нормального распределения с нужными параметрами  
sample50=func_list(50)
sigma=np.sqrt(2*k/50)
gucci=sts.norm(loc=k,scale=sigma)
pdf=gucci.pdf(x)
plt.plot(x,pdf)
plt.hist(sample50,normed=True)
plt.ylabel('y')
plt.xlabel('x')
plt.title('распределение выборочного среднего случайной величины при объёме выборки n=10')
```

Вывод:
* При увеличении размера выборки n, распределение средних значений начинает сходиться и становится похожим на нормальное распределение
